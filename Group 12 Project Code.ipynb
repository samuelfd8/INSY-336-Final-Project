### Another one of the issues that we noticed with google trends data is that there was a certain threshold at around 9 months where the data that would appear in the CSV files that we downloaded would be in weekly data instead of daily. What we did was to download two CSV's per team per season (except for the 23-24 season because the time frame was August to March 31st > 9 months). 

### This code was to make the txt file that we used for the datascraper. As you can see there was a certain construction of the google trends like that we input the teams names into as well as the start and end date of the different parts of the season, for the reasons mentioned above.

```python
import pandas as pd
from datetime import datetime
from dateutil.relativedelta import relativedelta

fixtures_df = pd.read_csv('premier_league_2021_2022_fixtures_stripped.csv')

fixtures_df['Date'] = pd.to_datetime(fixtures_df['Date'], errors='coerce')

fixtures_df_cleaned = fixtures_df.dropna(subset=['Date']).drop_duplicates()


fixtures_df_cleaned['DateTime'] = pd.to_datetime(fixtures_df_cleaned['Date'].dt.strftime('%Y-%m-%d') + ' ' + fixtures_df_cleaned['Time']) - pd.Timedelta(hours=5)
fixtures_df_cleaned.loc[(fixtures_df_cleaned['DateTime'].dt.month == 10) & (fixtures_df_cleaned['DateTime'].dt.day >= 31) | 
                        (fixtures_df_cleaned['DateTime'].dt.month == 11) & (fixtures_df_cleaned['DateTime'].dt.day <= 7), 'DateTime'] -= pd.Timedelta(hours=1)
fixtures_df_cleaned.loc[(fixtures_df_cleaned['DateTime'].dt.month == 3) & (fixtures_df_cleaned['DateTime'].dt.day >= 13) & 
                        (fixtures_df_cleaned['DateTime'].dt.day <= 27), 'DateTime'] += pd.Timedelta(hours=1)


fixtures_df_cleaned['Date'] = fixtures_df_cleaned['DateTime'].dt.strftime('%Y-%m-%d')
fixtures_df_cleaned['Time'] = fixtures_df_cleaned['DateTime'].dt.strftime('%H:%M')

fixtures_df_cleaned.drop(columns='DateTime', inplace=True)


def format_team_name_safe(team_name):
    if isinstance(team_name, str):
        return team_name.lower().replace(' ', '%20')
    else:
        return None


def generate_nine_month_urls(df, start_date):
    teams = pd.concat([df['Home'], df['Away']]).unique()
    team_urls = {}
    

    end_date_first_segment = start_date + relativedelta(months=8)
    start_date_second_segment = start_date + relativedelta(weeks=6)  
    end_date_second_segment = start_date_second_segment + relativedelta(months=8)

    for team in teams:
        formatted_team_name = format_team_name_safe(team)
        if formatted_team_name:
           
            urls = [
                f"https://trends.google.com/trends/explore?date={start_date.strftime('%Y-%m-%d')}%20{end_date_first_segment.strftime('%Y-%m-%d')}&q={formatted_team_name}",
                f"https://trends.google.com/trends/explore?date={start_date_second_segment.strftime('%Y-%m-%d')}%20{end_date_second_segment.strftime('%Y-%m-%d')}&q={formatted_team_name}"
            ]
            team_urls[team] = urls

    return team_urls

season_start_date = fixtures_df_cleaned['Date'].min()

all_team_nine_month_urls = generate_nine_month_urls(fixtures_df_cleaned, datetime.strptime(season_start_date, '%Y-%m-%d'))

for team, urls in all_team_nine_month_urls.items():
    print(f"{team}: {urls[0]}")
    print(f"{team}: {urls[1]}\n")

    

output_file_path = 'google_trends_urls_nine_month_segments.txt'

with open(output_file_path, 'w') as file:
    for team, urls in all_team_nine_month_urls.items():
        file.write(f"{team}: {urls[0]}\n")
        file.write(f"{team}: {urls[1]}\n\n") 

print(f"URLs written to {output_file_path}")

```

### NOTE: This is our code for the webscraper. For this to run, you must have Firefox and Geckodriver installed. Annotations are included to show what was done. Also, when the Firefox opens the first Google Trends page there is always a 429 error, however, if you just press refresh then the rest will work automatically.

```python
from selenium import webdriver
from selenium.webdriver.firefox.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException
import os
import time
import random

home_dir = os.path.expanduser('~')

custom_download_dir = os.path.join(home_dir, 'downloads/finalproject/2021-2022/csv') # This is our code to make sure that the CSVs is installed in an organized directory

os.makedirs(custom_download_dir, exist_ok=True)

"""
We decided to use firefox as our browser to do scraping on because it offered more privacy and a lower chance to be
detected by google than what we experienced when we used chrome drivers.
"""

firefox_options = Options()
firefox_options.set_preference("browser.download.folderList", 2)  
firefox_options.set_preference("browser.download.manager.showWhenStarting", False)
firefox_options.set_preference("browser.download.dir", custom_download_dir)
firefox_options.set_preference("browser.helperApps.neverAsk.saveToDisk", "text/csv")  

driver = webdriver.Firefox(options=firefox_options)

# This is our function for the scraper to find the download button on the CSV file page using the HTML code of the page
# We also used a randomized timer to make sure that the website did not notice that we were robots immediately. 
def download_csv(url):
    driver.get(url)
    time.sleep(random.uniform(5, 10))  
    try:
        download_button = WebDriverWait(driver, 10).until(
            EC.element_to_be_clickable((By.CSS_SELECTOR, 'button.widget-actions-item.export'))
        )
        download_button.click()
        time.sleep(random.uniform(5, 10))  # Delay after clicking to allow for download
    except TimeoutException:
        print("Timed out waiting for the download button to appear.")
    except Exception as e:
        print(f"An error occurred while trying to download CSV: {e}")

# this function goes through the file directory that we put the file in and renamed the file so it was easy for us to combine later
def wait_for_download_and_rename(download_dir, new_filename):
    print(f"Waiting for the download to complete...") # We added in these types of functions so we knew where the issue was if the whole code stopped working
    time.sleep(4)   
    while any([filename.endswith('.part') for filename in os.listdir(download_dir)]):
        time.sleep(1) 
    latest_file = max([download_dir + '/' + f for f in os.listdir(download_dir)], key=os.path.getctime)
    os.rename(latest_file, download_dir + '/' + new_filename)
    print(f"File renamed to {new_filename}")


# This reads URLs from the text file and download the CSV files
def process_link(team_name, url, period):
    download_csv(url)
    new_filename = f"{team_name}_{period}.csv"
    wait_for_download_and_rename(custom_download_dir, new_filename)

# This is the code that takes the txt file URLs and uses them
with open('google_trends_urls_nine_month_segments.txt', 'r') as file:
    lines = [line.strip() for line in file if line.strip()]  #Since there are white spaces between the links this strips them
    for i in range(0, len(lines), 2):  # This steps through two lines at a time for each team
        team_line = lines[i]
        team_name = team_line.split(':')[0].strip().replace(' ', '')
        
        #We decided to create two processes for the early and late links
        early_url = lines[i].split(':', 1)[1].strip()
        process_link(team_name, early_url, "early")
        
        # This processes "late" period link
        late_url = lines[i+1].split(':', 1)[1].strip()
        process_link(team_name, late_url, "late")
        
        time.sleep(random.uniform(15, 30))  # In between different ones we had a longer delay because when we had a smaller one the website would not let us download files anymore

driver.quit()
```

```python
fixtures_df=pd.read_csv('premier_league_2021_2022_fixtures.csv')
```

```python
# drop empty columns
fixtures=fixtures_df.dropna(subset=['Date'])
```

```python
#convert date to datetime format
fixtures.loc[:, 'Date']=pd.to_datetime(fixtures['Date'])
```

```python
#extract dates of all premier league games
unique_dates=np.sort(pd.unique(fixtures['Date']))
```

```python
#create dataframe with index of dates and columns of weeks
df=pd.DataFrame(index=unique_dates)
df.index=pd.to_datetime(df.index)
df['Week']=df.index.map(fixtures.set_index('Date')['Wk'].to_dict())
```

```python
#extract all team names
unique_teams=pd.unique(fixtures['Home'])
```

```python
#create team columns
for team in unique_teams:
    if team not in df.columns:
        df[team]=np.nan
```

```python
#add game results to each team column
for index, row in fixtures.iterrows():
    date=row['Date']
    home_team=row['Home']
    away_team=row['Away']
    score=row['Score']
    score=score.replace('–','-')
    score=score.replace('—','-')
    home_goals,away_goals=[int(x) for x in score.split('-')]
    if home_goals > away_goals:
        df.loc[date, home_team] = 3  
        df.loc[date, away_team] = 1  
    elif home_goals < away_goals:
        df.loc[date, home_team] = 1 
        df.loc[date, away_team] = 3 
    else:
        df.loc[date, home_team] = 2
        df.loc[date, away_team] = 2
```

```python
weeklyrank=pd.read_csv('Weekly rank data 21-22.csv')
```

```python
# create rank column for each team
for team in weeklyrank['21-22 Team'].unique():
    df[f'{team} Rank'] = np.nan
```

```python
# add the ranks of each team at the beginning of the week to the rank column
for index, row in weeklyrank.iterrows():
    current_week = row['21-22 Week']
    next_week = current_week + 1
    team = row['21-22 Team']
    rank = row['21-22 Rank']
    df.loc[df['Week'] == next_week, f'{team} Rank'] = rank
```

```python
# add each teams opponent for each week to an opponent column
opponents_dict = {}
for index, row in fixtures.iterrows():
    date=row['Date']
    home_team=row['Home']
    away_team=row['Away']
    if date not in opponents_dict:
        opponents_dict[date]={}
    opponents_dict[date][home_team]=away_team
    opponents_dict[date][away_team]=home_team
    
for team in unique_teams:
    df[f'{team} Opponent']=df.index.map(lambda x: opponents_dict.get(x, {}).get(team, np.nan))
```

```python
# finding out which team has the higher daily traffic index and their result in those games
higher_index = []
win_with_higher_index = []
loss_with_higher_index = []
tie_with_higher_index = []


for index, row in bigdf.iterrows():
    for team in teams:
        opponent_col = f"{team} Opponent"
        if opponent_col in row and pd.notna(row[opponent_col]):
            opponent = row[opponent_col]
            team_index = row[f"Daily Google Trends Data: {team}"]
            opponent_index = row[f"Daily Google Trends Data: {opponent}"]
            index_diff = team_index - opponent_index

            if index_diff > 0:  
                higher_index.append(team)
                if row[team] == 3:
                    win_with_higher_index.append(team)
                elif row[team] == 1:
                    loss_with_higher_index.append(team)
                else:  
                    tie_with_higher_index.append(team)

total_games_with_higher_index = len(higher_index)
wins_with_higher_index = len(win_with_higher_index)
losses_with_higher_index = len(loss_with_higher_index)
ties_with_higher_index = len(tie_with_higher_index)

win_rate_with_higher_index = wins_with_higher_index/ total_games_with_higher_index
loss_rate_with_higher_index = losses_with_higher_index/ total_games_with_higher_index
tie_rate_with_higher_index = ties_with_higher_index / total_games_with_higher_index

print(f"Win rate when having higher daily normalized traffic index: {win_rate_with_higher_index:.2f}")
print(f"Loss rate when having higher daily normalized traffic index: {loss_rate_with_higher_index:.2f}")
print(f"Tie rate when having higher daily normalized traffic index: {tie_rate_with_higher_index:.2f}")
```

```python
# finding out which team has the higher weekly average search traffic and their result in those games
for team in teams:
    traffic_column = f"Weekly Google Trends Traffic: {team}"
    bigdf[traffic_column] = pd.to_numeric(bigdf[traffic_column].astype(str).str.replace(',', ''), errors='coerce')

weekly_traffic = {}
for team in teams:
    weekly_traffic[team] = bigdf.groupby('Week')[traffic_column].mean()

weekly_traffic = {}
for team in teams:
    
    weekly_traffic[team]=bigdf.groupby('Week')[f"Weekly Google Trends Traffic: {team}"].mean()  
    bigdf[f"Weekly Google Trends Traffic: {team}"] = pd.to_numeric(bigdf[f"Weekly Google Trends Traffic: {team}"].astype(str).str.replace(',', ''), errors='coerce')

traffic_advantage = []
win_with_advantage = []
loss_with_advantage = []
tie_with_advantage = []

for index, row in bigdf.iterrows():
    current_week = row['Week']
    for team in teams:
        opponent_col = f"{team} Opponent"
        if opponent_col in row and pd.notna(row[opponent_col]):
            opponent = row[opponent_col]

            
            team_traffic = weekly_traffic[team].get(current_week, pd.NA)
            opponent_traffic = weekly_traffic[opponent].get(current_week, pd.NA)

            
            if pd.notna(team_traffic) and pd.notna(opponent_traffic):
                traffic_diff = team_traffic - opponent_traffic

                if traffic_diff > 0:
                    traffic_advantage.append(team)
                    if row[team] == 3:
                        win_with_advantage.append(team)
                    elif row[team] == 1:
                        loss_with_advantage.append(team)
                    else:
                        tie_with_advantage.append(team)

else:
    total_games_with_advantage = len(traffic_advantage)
    wins_with_advantage = len(win_with_advantage)
    losses_with_advantage = len(loss_with_advantage)
    ties_with_advantage = len(tie_with_advantage)

    win_rate_with_advantage = wins_with_advantage / total_games_with_advantage
    loss_rate_with_advantage = losses_with_advantage / total_games_with_advantage
    tie_rate_with_advantage = ties_with_advantage / total_games_with_advantage

    print(f"Win rate when having traffic advantage: {win_rate_with_advantage:.4f}")
    print(f"Loss rate when having traffic advantage: {loss_rate_with_advantage:.4f}")
    print(f"Tie rate when having traffic advantage: {tie_rate_with_advantage:.4f}")
```

The goal of this part of the analysis is to look at the relationship between the Google search traffic that a match and the 'competitiveness' of the match. By competitiveness, we esentially mean the quality of the two teams playing. To do this, we created a new variable called 'Sum of Ranks'. This variable added the rank of both teams in a match at the time of the match being played. So, in a table of 20 teams, were 1st is best and 20th is worst, the lower the Sum of Ranks, the higher the quality of the teams playing in the match. Clearly, this isn't always the case for 'competitiveness', since matches between bottom ranking teams are often more competitive than matches between high rank teams and low rank teams. By competitiveness, we really mean quality. For the case of search traffic, we created a variable called "Sum of Google Trends Traffic" that added the individual traffic of two teams together.

This is all to say that the objective is to build and graphs for each team characterized by:
X axis: the Sum of Ranks
Y-axis: the Sum of Google Trends Traffic

### Creating the plots and regressions for each team for this part of the analysis

The aim of this code is to create the plots and run the regressions for the teams specified in the teams_of_interest variable. This provides the output for our analysis on how different types of matches affect online engagement and performance.  First, the block of code runs a loop through each team specified in the teams_of_interest variable. It uses a list to store the data for each team and it goes through the dataframe to get the matches for the corresponding teams. It then extracts the columns for the specified team and the opponent team as well, and extracts their Weekly Google Search Traffic and their Weekly Rank. The code then aggregates the Weekly Google Search Traffic of both teams into sum_trends and the Weekly Rank of both teams into sum_ranks. Finally, a new dataframe is created called team_df which has these aforementioned columns of data. Lastly, subplots are created (scatter plots with trend lines) for each team and an OLS regression is performed alongside the summary of the regression. 

```python
teams_of_interest = ['Burnley', 'Watford', 'Norwich City',
                     'Leicester City', 'Leeds United', 'Southampton']  

n_rows = len(teams_of_interest) // 2 + len(teams_of_interest) % 2
n_cols = 2

plt.figure(figsize=(n_cols * 10, n_rows * 6))

for i, team in enumerate(teams_of_interest, start=1):
    plot_data = []  
    for index, row in df.iterrows():
        if pd.notna(row.get(team)) and row.get(team) in [1, 2, 3]:  # Check if the team played and there's a game result
            opponent_col = f'{team} Opponent'
            rank_col = f'{team} Rank'
            trends_col = f'Weekly Google Trends Traffic: {team}'

            opponent = row[opponent_col]
            opponent_rank_col = f'{opponent} Rank'
            opponent_trends_col = f'Weekly Google Trends Traffic: {opponent}'

            if pd.notna(row.get(rank_col)) and pd.notna(row.get(trends_col)):
                sum_ranks = row[rank_col]
                sum_trends = row[trends_col]

                if pd.notna(row.get(opponent_rank_col)) and pd.notna(row.get(opponent_trends_col)):
                    sum_ranks += row[opponent_rank_col]
                    sum_trends += row[opponent_trends_col]

                if sum_ranks > 0:
                    plot_data.append((sum_trends, sum_ranks))
    team_df = pd.DataFrame(plot_data, columns=['Sum of Google Trends Traffic', 'Sum of Ranks'])

    plt.subplot(n_rows, n_cols, i)
    plt.scatter(team_df['Sum of Google Trends Traffic'], team_df['Sum of Ranks'], alpha=0.5, label='Data Points')
    
    X = sm.add_constant(team_df['Sum of Google Trends Traffic'])  
    model = sm.OLS(team_df['Sum of Ranks'], X).fit()
    print(f"OLS Regression Results for {team}:\n", model.summary())
    trendline = model.predict(X)
    plt.plot(team_df['Sum of Google Trends Traffic'], trendline, color='purple', label='Trend Line')
    
    plt.title(f'{team}: Sum of Google Trends Traffic vs. Sum of Ranks')
    plt.xlabel('Sum of Google Trends Traffic')
    plt.ylabel('Sum of Ranks')
    plt.grid(True)
    plt.legend()

plt.tight_layout()  
plt.show()
```

## Analyzing the correlation between Google Trends data and team performance for both top and bottom teams


#### This part of the code was used to analyze the correlation between performance and google trends for top teams

```python
import pandas as pd
import numpy as np
# Load the dataset
file_path = 'the_holy_grail.csv'
data = pd.read_csv(file_path)
```

#### Adding a dataframe for the bottom 

```python
# Identifying columns for game results and Google Trends data
result_columns = [col for col in data.columns if 'Daily Google Trends Data:' not in col and col != 'Day']
trends_columns = [col for col in data.columns if 'Daily Google Trends Data:' in col]

# Creating a DataFrame for game results
game_results = data[['Day'] + result_columns].copy()
```

```python
# Creating a DataFrame for Google Trends data
google_trends = data[['Day'] + trends_columns].copy()
google_trends.columns = [col.replace('Daily Google Trends Data: ', '') for col in google_trends.columns]

# Combining game results with Google Trends data
combined_data = game_results.melt(id_vars=['Day'], var_name='Team', value_name='Result')
combined_data = combined_data.dropna(subset=['Result'])  # Removing days with no games
google_trends = google_trends.melt(id_vars=['Day'], var_name='Team', value_name='Google Trends')

# Merging on Day and Team
final_dataset = pd.merge(combined_data, google_trends, on=['Day', 'Team'])
```

```python
from scipy.stats import pointbiserialr
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

# Analysis for selected teams
teams_to_analyze = ['Chelsea', 'Manchester Utd', 'Manchester City', 'Arsenal', 'Tottenham', 'Liverpool']
analysis_results = {}

# Scaling the features
scaler = StandardScaler()

for team in teams_to_analyze:
    team_data = final_dataset[final_dataset['Team'] == team]
    X_team = team_data[['Google Trends']]
    y_team = team_data['Result'].astype(int)
    X_team_scaled = scaler.fit_transform(X_team)
    X_team_train, X_team_test, y_team_train, y_team_test = train_test_split(
        X_team_scaled, y_team, test_size=0.3, random_state=42)
    logistic_model_team = LogisticRegression(multi_class='multinomial', solver='lbfgs')
    logistic_model_team.fit(X_team_train, y_team_train)
    train_score_team = logistic_model_team.score(X_team_train, y_team_train)
    test_score_team = logistic_model_team.score(X_team_test, y_team_test)
    team_data['Win'] = (team_data['Result'] == 3).astype(int)
    correlation_team, p_value_team = pointbiserialr(team_data['Win'], team_data['Google Trends'])
    analysis_results[team] = {
        'Training Accuracy': train_score_team,
        'Test Accuracy': test_score_team,
        'Correlation Coefficient': correlation_team,
        'P-value': p_value_team
    }
    print(f"Analysis results for {team}:")
    print(analysis_results[team])
    print(" ")  
```

#### This part of the code was used to analyze the correlation between performance and google trends for bottom 2 teams

```python
file_path = 'the_holy_grail.csv'
data = pd.read_csv(file_path)

# Define bottom teams for each season
bottom_teams_2021_2022 = ['Norwich City', 'Watford', 'Burnley']
bottom_teams_2022_2023 = ['Southampton', 'Leicester City', 'Leeds United']

# Define columns based on team names for game results and Google Trends
result_columns_2021_2022 = bottom_teams_2021_2022
result_columns_2022_2023 = bottom_teams_2022_2023
trends_columns_2021_2022 = [f'Daily Google Trends Data: {team}' for team in bottom_teams_2021_2022]
trends_columns_2022_2023 = [f'Daily Google Trends Data: {team}' for team in bottom_teams_2022_2023]

# Combine columns for filtering
columns_2021_2022 = result_columns_2021_2022 + trends_columns_2021_2022 + ['Day']
columns_2022_2023 = result_columns_2022_2023 + trends_columns_2022_2023 + ['Day']

# Filter dataframes for each season
data_2021_2022 = data[columns_2021_2022]
data_2022_2023 = data[columns_2022_2023]

# Merging the Google Trends data with game results
data_2021_2022_melted = data_2021_2022.melt(id_vars=['Day'], var_name='Team', value_name='Value')
data_2022_2023_melted = data_2022_2023.melt(id_vars=['Day'], var_name='Team', value_name='Value')

# Split data into Google Trends and game results
trends_filter_2021_2022 = data_2021_2022_melted['Team'].str.startswith('Daily Google Trends Data:')
trends_data_2021_2022 = data_2021_2022_melted[trends_filter_2021_2022].copy()
trends_data_2021_2022['Team'] = trends_data_2021_2022['Team'].str.replace('Daily Google Trends Data: ', '')

game_results_2021_2022 = data_2021_2022_melted[~trends_filter_2021_2022].copy()

# Same for 2022-2023
trends_filter_2022_2023 = data_2022_2023_melted['Team'].str.startswith('Daily Google Trends Data:')
trends_data_2022_2023 = data_2022_2023_melted[trends_filter_2022_2023].copy()
trends_data_2022_2023['Team'] = trends_data_2022_2023['Team'].str.replace('Daily Google Trends Data: ', '')

game_results_2022_2023 = data_2022_2023_melted[~trends_filter_2022_2023].copy()

# Merge datasets for each season
final_data_2021_2022 = pd.merge(game_results_2021_2022, trends_data_2021_2022, on=['Day', 'Team'])
final_data_2022_2023 = pd.merge(game_results_2022_2023, trends_data_2022_2023, on=['Day', 'Team'])

# Combine the final datasets for both seasons
combined_final_data = pd.concat([final_data_2021_2022, final_data_2022_2023])

# Function to preprocess and analyze data
def combined_analysis(data):
    data = data.dropna(subset=['Value_x', 'Value_y']).copy()
    data['Win'] = (data['Value_x'] == 3).astype(int)
    scaler = StandardScaler()
    data['Google Trends Scaled'] = scaler.fit_transform(data[['Value_y']])
    X_train, X_test, y_train, y_test = train_test_split(
        data['Google Trends Scaled'].values.reshape(-1, 1), data['Win'], test_size=0.3, random_state=42
    )
    model = LogisticRegression()
    model.fit(X_train, y_train)
    train_accuracy = model.score(X_train, y_train)
    test_accuracy = model.score(X_test, y_test)
    correlation, p_value = pointbiserialr(data['Win'], data['Google Trends Scaled'])
    return {
        'Training Accuracy': train_accuracy,
        'Testing Accuracy': test_accuracy,
        'Correlation Coefficient': correlation,
        'P-value': p_value
    }

# Execute analysis for the combined dataset
combined_analysis_results = combined_analysis(combined_final_data)
combined_analysis_results
```
